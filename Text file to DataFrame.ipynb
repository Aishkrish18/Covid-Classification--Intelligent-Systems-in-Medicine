```python
imageNames = []

for file in os.listdir(path):
  imageNames.append(file)

df = pd.DataFrame(list(zip(imageNames)),columns=['Image Name'])
print(df)

#Store DataFrame as CSV File
df.to_csv('test_noisy.csv',index=False)
!cp test_noisy.csv "drive/My Drive/data/"
```

```python
readTxtFile = open(path + '/train.txt','r')
lines = readTxtFile.readlines()

count = 0
for line in lines:
  count+=1

print("\nTraining Images in Dataset: " + str(count))
```

```python
# Initialized empty lists
Normal = []
COVID = []
Pneumonia = []
Lung_Opacity = []
imageNames = []
imagePath = []

# Add Class Labels for Images and fix Image names
for line in lines:
  if('Normal' in line):
    Normal.append(1)
    imageNames.append(line.replace(' Normal','').replace('\n',''))
  else:
    Normal.append(0)

  if('COVID' in line):
    COVID.append(1)
    imageNames.append(line.replace(' COVID','').replace('\n',''))
  else:
    COVID.append(0)

  if('pneumonia' in line):
    Pneumonia.append(1)
    imageNames.append(line.replace(' pneumonia','').replace('\n',''))
  else:
    Pneumonia.append(0)

  if('Lung_Opacity' in line):
    Lung_Opacity.append(1)
    imageNames.append(line.replace(' Lung_Opacity','').replace('\n',''))
  else:
    Lung_Opacity.append(0)

# Create Image Path from Image Name
for img_name in imageNames:
  imagePath.append('/train/' + img_name)

## Lists of Corresponding Classes
#print(Normal)
#print(COVID)
#print(Pneumonia)
#print(Lung_Opacity)
#print(imageNames)
#print(imagePath)

# Class-wise distribution of images
print("Normal : " + str(sum(Normal)))
print("COVID : " + str(sum(COVID)))
print("Pneumonia : " + str(sum(Pneumonia)))
print("Lung Opacity : " + str(sum(Lung_Opacity)))
print("Total : " + str(sum(Normal)+sum(COVID) + sum(Pneumonia) + sum(Lung_Opacity)))
```

```python
#Convert above data into a data frame
df = pd.DataFrame(list(zip(imageNames,Normal,COVID,Pneumonia,Lung_Opacity,imagePath)),columns=['Image Name', 'Normal', 'COVID', 'Pneumonia', 'Lung_Opacity', 'Image Path'])
#df = pd.DataFrame(list(zip(imageNames,Normal,COVID,Pneumonia,Lung_Opacity)),columns=['Image Name', 'Normal', 'COVID', 'Pneumonia', 'Lung Opacity'])

#Store DataFrame as CSV File
#df.to_csv('train_with_labels.csv',index=False)
#!cp train_with_labels.csv "drive/My Drive/data/"
```

Create DataFrame with Image Name and Image Class
[Normal = 0, COVID = 1, Pneumonia = 2, Lung_Opacity = 3]

```python
imageClass = []

for index,row in df.iterrows():
  if(row['Normal']==1):
    appendClass = 0;
  elif(row['COVID']==1):
    appendClass = 1;
  elif(row['Pneumonia']==1):
    appendClass = 2;
  elif(row['Lung_Opacity']==1):
    appendClass = 3;
  imageClass.append(appendClass)

#print(imageClass)
#print(len(imageClass))
```

```python
df['Image Class'] = imageClass
```

```python
newDF = df[['Image Name', 'Image Class']].copy()

#Store DataFrame as CSV File
newDF.to_csv('ImageName_with_Class.csv',index=False)
!cp ImageName_with_Class.csv "drive/My Drive/data/"
```
